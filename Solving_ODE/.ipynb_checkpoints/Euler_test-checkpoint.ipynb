```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import Function, dsolve, Eq, Derivative, symbols
from sympy.abc import t
```

To solve
\begin{equation}
    A \ddot{x} + B\dot{x} + Cx = D
\end{equation}
using Euler's method, we need to split it into
\begin{equation}
    \dot{x} = v \\
    \dot{v} = \frac{1}{A}\left(D - Bv -Cx\right)
\end{equation}

```python
npoints = 1000
x = np.zeros(npoints)
v = np.zeros(npoints)
a = np.zeros(npoints)
A = 1
B = 1
C = 1
D = 1
dt = 0.1
time = np.linspace(0, 100, npoints)

for i in range(npoints-1):
    x[i+1] = x[i] + v[i]*dt
    v[i+1] = v[i] + (D - B*v[i] - C*x[i])/A * dt
```

```python
# Solves the differential equation analytically
y = Function('y')

ode = Eq(A*Derivative(y(t), t, t) + B*Derivative(y(t), t) + C*y(t), D)

# Solve the differential equation analytically
solution = dsolve(ode, y(t))

# Display the solution
print("General solution:")
print(solution)
```

```python
def exact_sol(C1, C2, t):
    return (C1*np.sin(np.sqrt(3)*t/2) + C2*np.cos(np.sqrt(3)*t/2))*np.exp(-t/2) + 1
```

Imposing BC $x(t) = 0$ and $\dot{x}(t) = 0$ we find that in our case $C_1 = -1/\sqrt{3}$ and $C_2 = -1$

```python
plt.figure()
plt.plot(time, x, label = 'numerical')
plt.plot(time, exact_sol(C1=-1/(np.sqrt(3)), C2=-1, t = time), label = 'analytical')
plt.legend()
plt.xlim(0,20)
plt.show()
```
