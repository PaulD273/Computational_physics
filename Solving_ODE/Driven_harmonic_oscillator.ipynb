A spring-mass system with friction is driven by an external force $F(t) = A \text{cos}(\omega t)$
\begin{equation}
    m \ddot{x} = -k x - \mu \dot{x} + F(t)\\
\end{equation}
Plot the frequency respone of this system.

Recasting this into a first order differential equation:
\begin{equation}
    \dot{x} = v \\
    \dot{v} = - \frac{k}{m} x - \frac{\mu}{m} v + A \text{cos}(\omega t)
\end{equation}

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
```

```python
def deriv(y, t, k, m, mu, A, omega):
    return np.array([y[1], -k/m * y[0] - mu/m * y[1] + A*np.cos(omega*t)])
```

```python
def rk4(y, t, k, m, mu, A, omega, dt):
    for i in range(len(t)-1):
        k1 = deriv(y[i][:], t[i], k, m, mu, A, omega) * dt
        k2 = deriv(y[i][:] + k1/2, t[i] + dt/2, k, m, mu, A, omega) * dt
        k3 = deriv(y[i][:] + k2/2, t[i] + dt/2, k, m, mu, A, omega) * dt
        k4 = deriv(y[i][:] + k3, t[i], k, m, mu, A, omega) * dt
        y[i+1][:] = y[i][:] + (k1 + 2*k2 + 2*k3 + k4)/6
    #return y
```

```python
m = 1  
mu = 1
A = 1
omega = 1
k = 1

t0 = 0
tf = 100
dt = 0.1
N = int((tf - t0)/dt)

y = np.zeros([N,2])
y_dot = np.zeros([N,2])

#y[0][0] = 1
#y[0][2] = 1

t = np.linspace(t0, tf, N)
```

```python
rk4(y, t, k, m, mu, A, omega, dt)
```

```python
np.transpose(y)
```

```python
x = np.transpose(y)[0]
v = np.transpose(y)[1]
```

```python
fig, (ax1, ax2) = plt.subplots(2,1)
ax1.plot(t, x)
ax1.set_ylabel('Displacement')
ax2.plot(t, v)
plt.xlabel('Time (s)')
ax2.set_ylabel('Velocity')
plt.show()
```
