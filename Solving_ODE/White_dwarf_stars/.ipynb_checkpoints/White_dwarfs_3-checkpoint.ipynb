```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import trapezoid
```

```python
def E_V(rho):
    x = (rho)**(1/3)
    return (x*(1+2*x**2)*np.sqrt(1+x**2) - np.log(x+np.sqrt(1+x**2)))
```

```python
def deriv(r, y):
    '''
    Equilibrium equation for white dwarfs
    '''
    rho, m = y
    gamma = get_gamma(rho)
    if gamma >= 0:
        return np.array([-m*rho/(gamma * r**2), rho*r**2])
    else:
        return 0
```

```python
def get_gamma(rho):
    '''
    Equation for the pressure
    '''
    if rho >= 0:
        x = (rho)**(1/3)
        gamma = x**2/(3*np.sqrt(1 + x**2))
    else:
        gamma = -1
    return gamma
```

```python
def rk4(t, deriv, y, dt, **kwargs):
    for i in range(len(t)-1):
        k1 = deriv(t[i], y[i][:], **kwargs) * dt
        k2 = deriv(t[i] + dt/2, y[i][:] + k1/2, **kwargs) * dt
        k3 = deriv(t[i] + dt/2, y[i][:] + k2/2,  **kwargs) * dt
        k4 = deriv( t[i], y[i][:] + k3, **kwargs) * dt
        y[i+1][:] = y[i][:] + (k1 + 2*k2 + 2*k3 + k4)/6
    #return y
```

```python
# Sets rho_c and the # of electrons per nucleon Y_e
rho_c = 1000
# For Fe(56) 
Y_e = 26/56
# For C(12)
#Y_e = 0.5
# Constats to set the dimensions into grams and centimeters
R0 = 7.72e8 * Y_e
M0 = 5.67e33 * Y_e**2
rho_0 = 9.79e5 * 1/Y_e
n0 = 5.89e29
# Mass of the electron
m_e = 9*10**(-34)
# Light' speed
c = 3e9
# Universal gravitational constant
G = 6.67*10**(-14)
# Mass and Radius of the sun
R_s = 6.95e10
M_s = 1.99e33

r0 = 0.01
rf = 3
dr = 0.001
N = int((rf - r0)/dr)

rho = np.zeros([N])
m = np.zeros([N])
y = np.zeros([N,2])


# The density at r=0 is rho_c, parametrized in input and the mass at r=0 is zero
y[0][0] = rho_c


r = np.linspace(r0, rf, N)
```

```python
# Vary rho_c
rho_c_list = np.arange(1, 100, 1)
radii = []
total_mass = []
fig, (ax1, ax2) = plt.subplots(2,1, figsize=(6, 8))
#plt.suptitle('Non Relativistic case')
fig.subplots_adjust(hspace=0.5)
ax1.set_xlabel(r'r/$R_0$')
ax1.set_ylabel(r'$\rho$/$\rho_0$')
ax2.set_xlabel(r'r/$R_0$')
ax2.set_ylabel(r'm/$M_0$')
for rho_c in rho_c_list:
    y[0][0] = rho_c
    rk4(r, deriv, y, dr)
    radii.append(r[np.argmin(np.transpose(y)[0])])
    total_mass.append(np.max(np.transpose(y)[1]))
    rho = np.transpose(y)[0]
    # Check that the density does not take negative values
    for i in range(N):
        if rho[i] <= 0:
            rho[i+1:] = 0
            break
    ax1.plot(r, np.transpose(y)[0], label = r'$\rho_{c}=$%s'% rho_c)
    ax2.plot(r, np.transpose(y)[1], label = r'$\rho_{c}=$%s'% rho_c)
#ax1.legend()
#ax2.legend()
ax1.set_xlim(0, 1)
ax2.set_xlim(0, 1)
plt.show()
```

```python
# Rescale in units of centimeters and grams
for i in range(len(radii)):
    radii[i] *= R0
radii
for i in range(len(total_mass)):
    total_mass[i] *= M0
```

```python
labels = ['Sirius B', 'Eri B', 'Stein 2051']
```
