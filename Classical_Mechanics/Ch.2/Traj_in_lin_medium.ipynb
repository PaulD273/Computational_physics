# Trajectory in a linear medium

```python
import numpy as np
import matplotlib.pyplot as plt
```

![ex_2.20.png](./ex_2.20.jpg)

$$
    x(t) = v_{x0} \tau \left ( 1 - e^{-t / \tau} \right ) = v_{0} \text{cos}\theta \tau \left( 1 - e^{-t / \tau} \right) \\
    y(t) = \left( v_{y0} + v_{\text{ter}} \right) \tau \left ( 1 - e^{-t / \tau} \right ) - v_{\text{ter}} t = \left( v_{0} \ \text{sin}\theta + v_{\text{ter}} \right) \tau \left ( 1 - e^{-t / \tau} \right ) - v_{\text{ter}} t
$$

```python
t = np.arange(0,3,0.001)
v0 = 1
m = 1
g = 1
theta = np.radians(45)
```

```python
def x_traj(t, v0, theta, m, b, **kwargs):
    if b == 0:
        tau = float('inf')
    else:
        tau = m/b
    return v0*tau*(1 - np.exp(-t/tau))
```

```python
def y_traj(t, v0, theta, m, b, **kwargs):
    if b == 0:
        tau = float('inf')
        vter = float('inf')
    else:
        tau = m/b
        vter = m*g/b
    return ((v0 + vter)*tau*(1 - np.exp(-t/tau)) - vter*t, vter)
```

```python
b_range = [1/0.3, 1.00, 1/1.3]
for b in b_range:
    x = x_traj(t, v0, theta, m, b)
    y, vter = y_traj(t, v0, theta, m, b)
    plt.plot(x,y, label = f'b = {round(b,2)}')
    plt.axvline(x = vter, linestyle = '--')
    plt.legend()
```

The stronger the drag, the smaller the range and the terminal velocity.

```python
t2 = np.arange(0,5,0.001)
b_range2 = [0.1, 0.05, 0.08]
for b in b_range2:
    x = x_traj(t2, v0, theta, m, b)
    y, vter = y_traj(t2, v0, theta, m, b)
    plt.plot(x,y, label = f'b = {round(b,2)}')
    plt.axvline(x = vter, linestyle = '--')
    plt.legend()
```
