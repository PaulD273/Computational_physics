```python
def cn(t, x, V, N, alpha):
    '''
    Crank-Nicolson algorithm
    V[j] column of values of u at a fixed time j
    '''
    a = np.ones([N])*2
    b = np.ones([N-1])*(-1)
    c = np.ones([N])*(1 + 2*alpha)
    A = np.diag(b, k=-1) + np.diag(a, k=0) + np.diag(b, k=1)
    B = np.diag(b*alpha, k=-1) + np.diag(c, k=0) + np.diag(b*alpha, k=1)
    for j in range(len(t)-1):
        V = np.linalg.inv(2*np.diag(np.ones([N])) + alpha*B)@(2*np.diag(np.ones([N])) - alpha*B) @ V 
```

```python
def rk4(t, deriv, y, dt, **kwargs):
    for i in range(len(t)-1):
        k1 = deriv(t[i], y[i][:], **kwargs) * dt
        k2 = deriv(t[i] + dt/2, y[i][:] + k1/2, **kwargs) * dt
        k3 = deriv(t[i] + dt/2, y[i][:] + k2/2,  **kwargs) * dt
        k4 = deriv( t[i], y[i][:] + k3, **kwargs) * dt 
        y[i+1][:] = y[i][:] + (k1 + 2*k2 + 2*k3 + k4)/6
    #return y
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
alpha = 1.0        # thermal diffusivity
L = 1.0            # length of the domain
T = 0.1            # total time
Nx = 20            # number of spatial steps
Nt = 100           # number of time steps

dx = L / Nx
dt = T / Nt
x = np.linspace(0, L, Nx+1)
t = np.linspace(0, L, Nt+1)
r = alpha * dt / dx**2

# Initial condition: u(x, 0) = sin(pi*x)
u = np.sin(np.pi * x)
u_new = np.zeros_like(u)

# Construct A and B matrices for Crank-Nicolson
main_diag = (1 + r) * np.ones(Nx-1)
off_diag = (-r / 2) * np.ones(Nx-2)
A = np.diag(main_diag) + np.diag(off_diag, 1) + np.diag(off_diag, -1)

main_diag = (1 - r) * np.ones(Nx-1)
B = np.diag(main_diag) + np.diag(-off_diag, 1) + np.diag(-off_diag, -1)

# Time stepping
for n in range(0, Nt):
    b = B @ u[1:-1]
    u_new[1:-1] = np.linalg.solve(A, b)
    u = u_new.copy()

# Plot result
plt.plot(x, u, label=f'Time = {T}')
plt.plot(x, np.sin(np.pi * x) * np.exp(-np.pi**2 * alpha * T), '--', label='Analytical')
plt.title('Crank-Nicolson: 1D Heat Equation')
plt.xlabel('x')
plt.ylabel('u(x,T)')
plt.legend()
plt.grid(True)
plt.show()
```
